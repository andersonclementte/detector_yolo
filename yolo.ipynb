```python
def dict_increment(dict1, key):
  if key in dict1.keys():
    dict1[key] = dict1[key] + 1 
  else:
    dict1[key] = 1

  return dict1
  
def print_class_counts(dict1, class_names):
  # print counts for each class name
  for key, val in dict1.items():
    print(class_names[key], val)

def count_vehicles(detection_res, confidence_threshold=0.5):
  counts = dict()
  # print(res.names.index('car'), res.names.index('bus'), res.names.index('truck'))

  for pred in detection_res.xyxyn[0]:
    confidence = pred[-2]
    if confidence > confidence_threshold:
      # print(pred)

      class_id = int(pred[-1])
      counts = dict_increment(counts, class_id)

  # print_class_counts(counts, detection_res.names)
  return counts
```

```python
import cv2
from IPython.display import clear_output

# Load YOLOv5 model
model_highway = yolov5.load('DLTrafficCounter/models/yolov5s_highway.pt', device='cuda:0')


video_path = 'traffic.mp4'

# mp4 = open(video_path, "rb").read()
# data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
# HTML(f"""""")
capture = cv2.VideoCapture(video_path)
film = []
vehicle_counts = []
while True:
  grabbed, frame = capture.read()
  if not grabbed:
        # print("End of video")
        break
  detected = model_highway(frame)
  detected.save("results")

  clear_output(wait=True)
  # count_vehicles(detected)
  counts = count_vehicles(detected)
  vehicle_counts.append(counts)
  # cv2.imshow(detected.ims[0])
  film.append(detected.ims[0])
  # print(dir(detected))
  # print(grabbed, frame.shape)
  # time.sleep(5)
print(len(film))
```

```python
def display_frame(film, frame_index):
    # Check if the index is within the bounds of the 'film' list
    if 0 <= frame_index < len(film):
        # Display the frame using OpenCV
        cv2.imshow('Frame', film[frame_index])

        # Wait for a key press to close the window
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    else:
        print(f"Frame index {frame_index} is out of bounds. There are {len(film)} frames in the list.")
```

```python
display_frame(film, 450)
```

```python
def save_video(film, output_path='output.mp4', frame_rate=24):
    # Get frame dimensions
    height, width = film[0].shape[:2]

    # Define the codec and create a VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Using 'mp4v' codec for MP4 format
    out = cv2.VideoWriter(output_path, fourcc, frame_rate, (width, height))

    for frame in film:
        # Convert frame to the correct color format if necessary (BGR to RGB)
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        out.write(frame)

    # Release the VideoWriter object
    out.release()
```

```python
save_video(film)
```
